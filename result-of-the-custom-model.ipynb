```python
import tensorflow as tf
from transformers import AutoConfig,TFAutoModelForPreTraining,AutoTokenizer

# Making the model
MODEL="gpt2"

# Adding the special tokens for the later use

SPECIAL_TOKENS  = { "bos_token": "<|BOS|>",
                    "eos_token": "<|EOS|>",
                    "unk_token": "<|UNK|>",                    
                    "pad_token": "<|PAD|>",
                    "sep_token": "<|SEP|>"}

# Adding the maxlength
MAXLEN          = 768  #{768, 1024, 1280, 1600}

# Function to make the tokenizer and adding special tokens to it as well.
def get_tokenizer(special_tokens=None):
    tokenizer = AutoTokenizer.from_pretrained(MODEL,return_tensors="tf") #GPT2Tokenizer

    if special_tokens:
        tokenizer.add_special_tokens(special_tokens)
        print("Special tokens added")
    return tokenizer

tokenizer=get_tokenizer(special_tokens=SPECIAL_TOKENS)

#GPT2LMHeadModel
config = AutoConfig.from_pretrained(MODEL, 
                                    bos_token_id=tokenizer.bos_token_id,
                                    eos_token_id=tokenizer.eos_token_id,
                                    output_hidden_states=False)

#----------------------------------------------------------------#
model = TFAutoModelForPreTraining.from_pretrained(MODEL, config=config)

#Special tokens added, model needs to be resized accordingly
model.resize_token_embeddings(len(tokenizer))

# Adding the model state now
model.load_weights("../input/finetunedmodelinternship/finetunedmodel/tf_model.h5")
```

```python
try :
    title = input("Enter the title")
    keywords = input("Enter the keywords followed with (,)")
    text = input("Enter the initial text")
except :
    title = "deep_learning"
    keywords = "nlp"
    text = "This model tells us about "
prompt = SPECIAL_TOKENS['bos_token'] + title + SPECIAL_TOKENS['sep_token'] + keywords + SPECIAL_TOKENS['sep_token']+text +SPECIAL_TOKENS['eos_token']
encoded_text = tokenizer.encode(prompt)
encoded_text = tf.convert_to_tensor([encoded_text])
sample_output = model.generate(encoded_text, 
                                do_sample=True,   
                                min_length=50, 
                                max_length=MAXLEN,
                                top_k=30,                                 
                                top_p=0.7,        
                                temperature=0.9,
                                repetition_penalty=2.0,
                                num_return_sequences=10)
for i, sample in enumerate(sample_output):
    text = tokenizer.decode(sample, skip_special_tokens=True)
    a = len(text)   
    print(text)
    print("{}: {}\n\n".format(i+1,  text[a:]))
```
